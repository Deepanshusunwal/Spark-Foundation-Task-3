# Exploratory Data Analysis

```python
data['State'].value_counts().head()
```

```python
#State vs number of items sold plot
plt.figure(figsize=(15,12))
sns.countplot(x='State',data=data,palette='rocket',order=data['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
data.hist(figsize=(10,10),bins=50)
plt.show()
```

```python
# Most customers tends to buy quantity of 2 and 3
# Discount give maximum is 0 to 20 percent
#Pie Chart Showing the % of sales in all the regions
plt.figure(figsize=(10,8))
data['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

# Profit vs Discount

```python
#Scatter plot for plotting the profit when a particular item is sold
fig,ax=plt.subplots(figsize=(20,8))
ax.scatter(data['Sales'],data['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=data)
plt.legend()
plt.show()
```

# Profit vs Quantity

```python
#No correlation between profit and discount
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=data)
plt.legend()
plt.show()
```

```python
data.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['red','green'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
#Profit and sales are maximum in consumer segment and minimum in Home Office segment
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=data,hue='Segment',order=data['Region'].value_counts().index,palette='cubehelix')
plt.xlabel('Region',fontsize=15)
plt.show()
```

```python
#Segment wise sales are more in east adn west region compared to central and south region
data.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['orange','yellow'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
#Profit and sales are maximum in west region and minimum in south region
ps = data.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['purple','green'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

```python
#highest profit is from california and new york, losses are from texas, pennsylvania, Ohio
t_states = data['State'].value_counts().nlargest(10)
t_states
```

```python
data.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['brown','cyan'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
ps = data.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','lightblue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```
